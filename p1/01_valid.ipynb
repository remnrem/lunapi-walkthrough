## Save the sample list

```python
sl = proj.sample_list()
sl
```

To facilitate other notebooks we'll save a text sample list (i.e. so they can load the same sample-list easily - although in principle one can also easily `build()` the lunapi sample list instead).    Before doing this, we'll need to strip out some special characters from the annotations field (the `{` and `}` characters above, as the annotation column can reflect a _set_ of annotation files, not just one).

```python
sl['Annotations'] = sl['Annotations'].map( lambda x: str(x).strip('{}\'') )
sl
```

Write the sample list to a file, `s1.lst`:

```python
sl.to_csv( 's1.lst', sep="\t", index=None, header=False)
```
